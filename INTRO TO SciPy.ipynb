```python
import numpy as np
from scipy import optimize
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def test_func(x, a, b):
    return a * np.sin(b*x)
```

```python
#seed the random number generator
np.random.seed(0)
```

```python
#generate points
x_data = np.linspace(-5, 5, num = 50)
y_data = 2.9 * np.sin(1.5 * x_data) + np.random.normal(size=50)
```

```python
params, covariance = optimize.curve_fit(test_func, x_data, y_data, p0=[2,2])
```

```python
#finally,plot the values
plt.figure(figsize = [6,4])
plt.scatter(x_data, y_data, label='DATA')
plt.plot(x_data, test_func(x_data, params[0],
                          params[1]), label='Fitted function')
plt.legend(loc='best')
```

```python

xx_data = np.linspace(2.9, 14, num = 50)
yy_data = 2.9 * np.sin(1.5 * xx_data) + np.random.normal(size=50)

params, covariance = optimize.curve_fit(test_func, xx_data, yy_data, p0=[2,2])

#finally,plot the values
plt.figure(figsize = [6,4])
plt.scatter(xx_data, yy_data, label='DATA')
plt.plot(x_data, test_func(xx_data, params[0],
                          params[1]), label='Fitted function')
plt.legend(loc='best')
```

```python
xx_data = np.linspace(-5, 5, num = 50)
yy_data = 2.9 * np.cos(1.5 * xx_data) + np.random.normal(size=50)

params, covariance = optimize.curve_fit(test_func, xx_data, yy_data, p0=[2,2])

#finally,plot the values
plt.figure(figsize = [6,4])
plt.scatter(xx_data, yy_data, label='DATA')
plt.plot(x_data, test_func(xx_data, params[0],
                          params[1]), label='Fitted function')
plt.legend(loc='best')
```

```python
np.random.seed(0)

# generate sparse data
measured_time = np.linspace(0, 1, 10)
noise = 1e-1 * (np.random.random(10) * 2 - 1)
measures = np.sin(2 * np.pi * measured_time) + noise

```

```python
# interpolate to provide new time points
interpolation_time = np.linspace(0, 1, 50)

linear_interp = interp1d(measured_time, measures)
linear_results = linear_interp(interpolation_time)

cubic_interp = interp1d(measured_time, measures, kind='cubic')
cubic_results = cubic_interp(interpolation_time)

```

```python
#plot the data and the interpolations
plt.figure(figsize=[6, 4])
plt.plot(measured_time, measures, 'o', ms=6, label='measures')
plt.plot(interpolation_time, linear_results, label='linear interpolation')
plt.plot(interpolation_time, cubic_results, label='cubic interpolation')
plt.legend()
```
